* Doing bi-variate analysis on numerical values.

```python
numerical_cols = ['price', 'mileage', 'tax', 'mpg', 'engineSize']
matrix= df[numerical_cols].corr()
```

```python
matrix
```

```python
sns.heatmap(matrix , annot = True, cmap = 'coolwarm')
plt.title(('correation of numerical features'.upper()))
plt.show()
```

 * Price and year have a positive correlation (newer cars  higher price).
* Mileage and price have a strong negative correlation (higher mileage cheapercars).

----------------------------------------------------------------------------------------------------------------------------------------------------------------------
* Doing bi-variate analysis on numerical with categorical values.

```python
plt.figure(figsize=(8,6))
sns.barplot(
    x='fuelType', 
    y='price', 
    hue='transmission', 
    data=df, 
    palette='viridis', 
    errorbar=None
)
plt.title('Average BMW Price by Fuel Type and Transmission')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.show()
```

* Automatic petrol BMWs have the highest average price, while manual diesels are generally the cheapest, showing how both fuel type and transmission influence car value.

```python
sns.scatterplot(data = df , x='mileage',y='price',hue = 'transmission')
```

* Most high end cars are either semi-automatic or automatic and have very bad mileage.
* expensive cars have bad mileage .
* most of the affordable cars are automatic followed by manual.
* manual car are more affordable and have better mileage than other transmission.

```python
sns.boxplot(
    x='transmission',
    y='price',
    hue='fuelType',
    data=df,
    palette='Set1', # A nice color palette
    whis=[5, 95]       # Optional: Define whiskers to show 5th and 95th percentiles (instead of default 1.5*IQR)
)
plt.show()
```

* The box plot visualizes car price distribution (y) by Transmission (x), further split by Fuel Type (hue). The points shown beyond the lines (whiskers) are outliers, representing prices that fall outside the central 90% range of cars for that specific transmission and fuel type combination.

```python
plt.figure(figsize= (30,14))
sns.countplot(data = df, x='model')
plt.xlabel('Model')
plt.ylabel('NUMBER OF CARS LISTED')
```

* 1 & 3 series BMW are the most listed model over the period.
* M AND Z series BMW  are lowest listed because it is their high end models.
